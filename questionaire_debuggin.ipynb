```python
swbs_data = df[["Q53"]+Utils.SWBS_COLS].drop(index=[0, 1])
swbs_data = swbs_data.dropna()
Social_Integration = ["Q90_2", "Q90_6", "Q90_11"]
Social_Acceptance = ["Q90_3", "Q90_10", "Q90_14"]
Social_Contribution = ["Q90_4", "Q90_7", "Q90_15"]
Social_Actualization = ["Q90_5", "Q90_9", "Q90_13"]
Social_Coherence = ["Q90_1", "Q90_8", "Q90_12"]
swbs_data["Social Integration"] = swbs_data[Social_Integration].sum(axis=1)
swbs_data["Social Acceptance"] = swbs_data[Social_Acceptance].sum(axis=1)
swbs_data["Social Contribution"] = swbs_data[Social_Contribution].sum(axis=1)
swbs_data["Social Actualization"] = swbs_data[Social_Actualization].sum(axis=1)
swbs_data["Social Coherence"] = swbs_data[Social_Coherence].sum(axis=1)

swbs_data
```

```python
ric_data = df[["Q53"] + Utils.RIC_COLS].drop(index=[0, 1])
ric_data = ric_data.dropna()
ric_data["RiC Q1 Score"] = ric_data.iloc[:, 1]
ric_data["RiC Q2 Score"] = ric_data.iloc[:, 2]
frequency = ["Q3","Q4","Q5","Q6","Q7","Q8"]
same_freq = ["Q3","Q5","Q7"]
diff_freq = ["Q4","Q6","Q8"]

quality = ["Q3a","Q4a","Q5a","Q6a","Q7a","Q8a"]
same_qual = ["Q3a","Q5a","Q7a"]
diff_qual = ["Q4a","Q6a","Q8a"]

ric_data["Frequency"] = ric_data[frequency].mean(axis=1)
ric_data["Quality"] = ric_data[quality].mean(axis=1)
ric_data["Same Race Frequency"] = ric_data[same_freq].mean(axis=1)
ric_data["Different Race Frequency"] = ric_data[diff_freq].mean(axis=1)
ric_data["Same Race Quality"] = ric_data[same_qual].mean(axis=1)
ric_data["Different Race Quality"] = ric_data[diff_qual].mean(axis=1)
```
